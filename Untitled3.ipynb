# Credit Risk Modeling: PD, LGD, and EAD
This notebook performs a comprehensive credit risk analysis including data preprocessing, EDA, feature engineering, and model development for Probability of Default (PD), Loss Given Default (LGD), and Exposure at Default (EAD).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = '/content/Loan payments data.csv'
df = pd.read_csv(file_path)

# Display basic info and first few rows
print("Dataset Info:")
display(df.info())
print("\nFirst 5 rows:")
display(df.head())

# Summary statistics
print("\nSummary Statistics:")
display(df.describe(include='all'))
```

### 1. Dataset Definition and Understanding

Based on the initial load, we need to define our targets:
- **PD Target**: Usually derived from `loan_status` (e.g., 'PAIDOFF' vs 'COLLECTION'/'COLLECTION_PAIDOFF').
- **LGD Target**: Relates to the amount lost after default. We will look at `past_due_days` and principal amounts to estimate recovery.
- **EAD Target**: The total exposure at the moment of default.

Let's check the unique values in `loan_status` to define default.

```python
print("Loan Status counts:")
display(df['loan_status'].value_counts())

# Define Default (Binary for PD)
df['is_default'] = df['loan_status'].apply(lambda x: 1 if x in ['COLLECTION', 'COLLECTION_PAIDOFF'] else 0)

print("\nTarget Variable (PD) Distribution:")
display(df['is_default'].value_counts(normalize=True))
```

### 2. Data Preprocessing
We need to:
- Convert date columns to datetime objects.
- Handle missing values (especially in `paid_off_time`).
- Encode categorical variables (`education`, `Gender`).
- Define LGD and EAD targets.

```python
# 1. Date Conversion
date_cols = ['effective_date', 'due_date', 'paid_off_time']
for col in date_cols:
    df[col] = pd.to_datetime(df[col])

# 2. Define LGD and EAD
# EAD: Total amount at risk at time of default (Principal)
df['EAD'] = df['Principal']

# LGD: (EAD - Recovery) / EAD.
# For this dataset, we'll assume COLLECTION_PAIDOFF recovers some amount,
# while COLLECTION recovers less. Let's simplify:
# COLLECTION = 100% loss, COLLECTION_PAIDOFF = 50% loss, PAIDOFF = 0% loss.
def calculate_lgd(row):
    if row['loan_status'] == 'COLLECTION':
        return 1.0
    elif row['loan_status'] == 'COLLECTION_PAIDOFF':
        return 0.5
    else:
        return 0.0

df['LGD'] = df.apply(calculate_lgd, axis=1)

# 3. Categorical Encoding
df['Gender'] = df['Gender'].map({'male': 0, 'female': 1})
df = pd.get_dummies(df, columns=['education'], drop_first=True)

print("Preprocessed Data Info:")
display(df.info())
display(df.head())
```

### 3. Exploratory Data Analysis (EDA)
Let's visualize the relationship between features and our targets.

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.countplot(x='terms', hue='is_default', data=df)
plt.title('Default Rate by Loan Terms')

plt.subplot(1, 3, 2)
sns.boxplot(x='is_default', y='age', data=df)
plt.title('Age Distribution vs Default')

plt.subplot(1, 3, 3)
sns.histplot(df['LGD'], bins=10, kde=False)
plt.title('LGD Distribution')

plt.tight_layout()
plt.show()
```

### 4. Feature Engineering and Data Splitting
We will prepare the feature set and split the data into training (65%) and testing (35%) sets.

```python
from sklearn.model_selection import train_test_split

# Selecting features for the models
# We'll use demographic data and loan characteristics
features = ['Principal', 'terms', 'age', 'Gender',
            'education_High School or Below', 'education_Master or Above', 'education_college']

X = df[features]
y_pd = df['is_default']
y_lgd = df['LGD']
y_ead = df['EAD']

# Perform 65/35 Split
X_train, X_test, y_pd_train, y_pd_test, y_lgd_train, y_lgd_test, y_ead_train, y_ead_test = train_test_split(
    X, y_pd, y_lgd, y_ead, test_size=0.35, random_state=42
)

print(f"Training set size: {X_train.shape[0]}")
print(f"Test set size: {X_test.shape[0]}")
```

### 5. & 6. Model Development and Evaluation
We will now build the PD, LGD, and EAD models and evaluate them using appropriate metrics.

```python
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.metrics import roc_auc_score, confusion_matrix, mean_squared_error, r2_score

# --- PD Model (Logistic Regression) ---
pd_model = LogisticRegression()
pd_model.fit(X_train, y_pd_train)
pd_probs = pd_model.predict_proba(X_test)[:, 1]
pd_preds = pd_model.predict(X_test)

# --- LGD Model (Linear Regression) ---
# LGD is typically modeled only on defaults, but for this simplified exercise,
# we will use the full set to predict potential loss rates.
lgd_model = LinearRegression()
lgd_model.fit(X_train, y_lgd_train)
lgd_preds = lgd_model.predict(X_test)

# --- EAD Model (Linear Regression) ---
ead_model = LinearRegression()
ead_model.fit(X_train, y_ead_train)
ead_preds = ead_model.predict(X_test)

# Evaluation
print("--- PD Model Evaluation ---")
print(f"AUC-ROC: {roc_auc_score(y_pd_test, pd_probs):.4f}")
print("Confusion Matrix:")
print(confusion_matrix(y_pd_test, pd_preds))

print("\n--- LGD Model Evaluation ---")
print(f"RMSE: {np.sqrt(mean_squared_error(y_lgd_test, lgd_preds)):.4f}")

print("\n--- EAD Model Evaluation ---")
print(f"RMSE: {np.sqrt(mean_squared_error(y_ead_test, ead_preds)):.4f}")
```

```python
from sklearn.metrics import roc_curve

# Calculate ROC curve values
fpr, tpr, thresholds = roc_curve(y_pd_test, pd_probs)

# Plot ROC Curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc_score(y_pd_test, pd_probs):.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) - PD Model')
plt.legend(loc='lower right')
plt.grid(alpha=0.3)
plt.show()
```

### 10. Model Improvement: Scaling and Random Forest
The initial Logistic Regression model struggled with convergence and provided a modest AUC. We will now:
1. Scale the numerical features.
2. Implement a Random Forest Classifier to capture non-linear patterns.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier

# 1. Feature Scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 2. Re-run Logistic Regression with scaled data
log_reg_scaled = LogisticRegression(max_iter=1000)
log_reg_scaled.fit(X_train_scaled, y_pd_train)
probs_log_scaled = log_reg_scaled.predict_proba(X_test_scaled)[:, 1]

# 3. Implement Random Forest
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_pd_train) # Tree-based models don't strictly require scaling
probs_rf = rf_model.predict_proba(X_test)[:, 1]

print(f"Logistic Regression (Scaled) AUC: {roc_auc_score(y_pd_test, probs_log_scaled):.4f}")
print(f"Random Forest AUC: {roc_auc_score(y_pd_test, probs_rf):.4f}")
```

### 7. Model Interpretation
Let's examine the coefficients of the PD and LGD models to understand the risk drivers.

```python
# PD Model Coefficients
pd_coefs = pd.DataFrame({'Feature': features, 'Coefficient': pd_model.coef_[0]})
print("PD Model Coefficients (Log-Odds):")
display(pd_coefs.sort_values(by='Coefficient', ascending=False))

# LGD Model Coefficients
lgd_coefs = pd.DataFrame({'Feature': features, 'Coefficient': lgd_model.coef_})
print("\nLGD Model Coefficients:")
display(lgd_coefs.sort_values(by='Coefficient', ascending=False))
```

### 8. Expected Loss (EL) Estimation
Expected Loss is calculated as: `EL = PD * LGD * EAD`.

```python
# Calculate Expected Loss on the test set
# pd_probs contains the probability of default
# lgd_preds contains the predicted loss rate
# ead_preds contains the predicted exposure

el_results = pd.DataFrame({
    'Actual_Default': y_pd_test,
    'PD_Prob': pd_probs,
    'LGD_Pred': lgd_preds,
    'EAD_Pred': ead_preds
})

el_results['Expected_Loss'] = el_results['PD_Prob'] * el_results['LGD_Pred'] * el_results['EAD_Pred']

print("Expected Loss Summary (First 10 rows of Test Set):")
display(el_results.head(10))

print(f"\nTotal Portfolio Expected Loss (Test Set): ${el_results['Expected_Loss'].sum():.2f}")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(el_results['Expected_Loss'], bins=20, kde=True, color='teal')
plt.title('Distribution of Expected Loss (Test Set)')
plt.xlabel('Expected Loss ($)')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualize PD, EAD, and LGD distributions
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

sns.histplot(el_results['PD_Prob'], bins=20, kde=True, ax=axes[0], color='blue')
axes[0].set_title('PD (Probability of Default) Distribution')

sns.histplot(el_results['EAD_Pred'], bins=20, kde=True, ax=axes[1], color='green')
axes[1].set_title('EAD (Exposure at Default) Distribution')

sns.histplot(el_results['LGD_Pred'], bins=20, kde=True, ax=axes[2], color='red')
axes[2].set_title('LGD (Loss Given Default) Distribution')

plt.tight_layout()
plt.show()
```

### Identification of Riskiest and Healthiest Loans
We define 'Riskiest' as loans with the highest Expected Loss and 'Healthiest' as those with the lowest.

```python
# Identify Riskiest and Healthiest Loans
riskiest_loans = el_results.sort_values(by='Expected_Loss', ascending=False).head(5)
healthiest_loans = el_results.sort_values(by='Expected_Loss', ascending=True).head(5)

print("Top 5 Riskiest Loans (Highest Expected Loss):")
display(riskiest_loans)

print("\nTop 5 Healthiest Loans (Lowest Expected Loss):")
display(healthiest_loans)
```
